# Implemente uma função Python para cada uma das seguintes métricas de regressão:


*   MAE
*   MSE
*   RMSE

 Cada função deve receber valores reais (y) e valores preditos (y_pred) como parâmetros de entrada e retornar o valor calculado da métrica. Esse parâmetros podem ser arrays do numpy ou DataFrames do Pandas. O ideal é realizar as operações de forma vetorizada, evitando acessar cada célula individualmente. Não usar as implementações já existentes do Scikit-Learn.






```python
import pandas as pd
import numpy as np
```

```python
y = [1,4]
t_pred = [3,4]
```

```python
def calculateMAE(y_true,y_pred):
   mae=np.sum(np.absolute(y_true - y_pred)) / y_true.shape[0]
   return mae
```

```python
def calculateMSE(y_true,y_pred):
   mse=np.absolute(np.subtract(y_true, y_pred)**2).mean()
   return mse
```

```python
def calculateRMSE(y_true,y_pred):
   rmse=np.sqrt(np.absolute(np.subtract(y_true, y_pred)**2).mean())
   return rm
```

```python
mae = calculateMAE(y,t_pred)
mse = calculateMSE(y,t_pred)
rm = calculateRMSE(y,t_pred)
print("erro medio absoluto é : ",mae)
print("erro quadrático médio ",mse)
print("Raiz quadrada do erro-médio é aproximadamente : ",rm)
```
